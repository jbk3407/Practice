```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (3,3)
```

```python
#ex 12-1
data = [20,30,40]
x=[1,2,3]

plt.plot(x,data)
plt.show()
```

```python
#ex 12-2

plt.bar(['kim', 'lee', 'kang'], [85,88,90])

plt.title('English Score of three students')
plt.xlabel('Student name')
plt.ylabel('Scores')
plt.show()
```

```python
#ex 12-3

#한글 처리하는 방식
import matplotlib as mlp
import matplotlib.pyplot as plt
mlp.rcParams['axes.unicode_minus']=False
mlp.rc('font', family = 'Malgun Gothic')

font1 = {'size':18, 'color':'green'}

xdata = ['안지영', '홍지수', '황예린']
ydata = [90,85,88]

plt.plot(xdata, ydata)

plt.title("세명의 학생의 국어 성적", fontdict=font1)
plt.xlabel('이름')
plt.ylabel('성적')

plt.show()
```

```python
#ex 12-4

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90,85,88]
ydata2 = [83, 88, 91]

plt.plot(xdata, ydata1, label='국어')
plt.plot(xdata, ydata2, label='엉어')

plt.legend(loc = 'upper center')     #범례
plt.title('세명 학생의 국어, 영어 성적')

plt.show()
```

```python
#ex 12-5 Line style

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90,85,88]
ydata2 = [83, 100, 91]
ydata3 = [85, 20, 56]
ydata4 = [56, 46, 90]

plt.plot(xdata, ydata1, label='국어', color='red', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label='엉어', color='#00ffff', linestyle='--', marker='x')
plt.plot(xdata, ydata3, label='수학', color='magenta', linestyle='-.', marker='s')
plt.plot(xdata, ydata4, label='사회', color='#444444', linestyle=':', marker='d')

plt.title('세명 학생의 네과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
#ex 12-5 Line style

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90,85,88]
ydata2 = [83, 100, 91]
ydata3 = [85, 20, 56]
ydata4 = [56, 46, 90]

plt.plot(xdata, ydata1, 'ro-', label='국어', linewidth=3, markersize=15)  # 'ro-' = 'color,marker,style'
plt.plot(xdata, ydata2, 'cx--', label='엉어', linewidth=3, markersize=15)
plt.plot(xdata, ydata3, 'ms--', label='수학', linewidth=3, markersize=15)
plt.plot(xdata, ydata4, 'kd:', label='사회', linewidth=3, markersize=15)

plt.title('세명 학생의 네과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
#ex 12-7

mlp.rc('font', family='Malgun Gothic')
x=['영희', '철수', '재호']
y=[3,1,5]

plt.title('연간 영화관람 회수')
plt.bar(x,y)

plt.ylabel('회수')
plt.show()
```

```python
#시그모이드
import numpy as np
xs = np.linspace(-3,3,601)
ys = 1/(1+np.exp(-xs))

plt.plot(xs,ys)
plt.ylim(-0.1, 1.1)
plt.title('시그모이드 그래프')
plt.axhline(y=0, color='k', linewidth='0.5') #수평선 축(x축)
plt.axhline(y=1, color='k', linewidth='0.5')
plt.axvline(x=0, color='k', linewidth='0.5')  #수직선 축(y축)

plt.show()
```
