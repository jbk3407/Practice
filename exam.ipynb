```python

from random import *

temp=str(input('가로크기(size_x) 세로크기(size_y)....')).split(' ')
size_x,size_y=int(temp[0]),int(temp[1])

def create_mine(n):                   #매트릭스를 출력해주는 함수
    for i in range (0,size_y):
        for j in range (0,size_x):
            print(n[i][j],end=' ')
        print()

def check_mat(n):                   #매트릭스를 체크하여 주변의 지뢰 갯수를 세어주는 함수
    for i in range (0,size_y):
        for j in range (0,size_x):
            if n[i][j]!='*':
                if j==0:
                    x0=0
                else:
                    x0=j-1

                if j==size_x-1:
                    x1=size_x-1
                else:
                    x1=j+1

                if i==0:
                    y0=0
                else:
                    y0=i-1

                if i==size_y-1:
                    y1=size_y-1
                else:
                    y1=i+1

                mine_sum=0
                for yy in range (y0, y1+1):
                    for xx in range (x0, x1+1):
                        if n[yy][xx]=='*':
                            mine_sum+=1

                n[i][j]=mine_sum

mat=[]
for i in range (0,size_y):          
    temp=[]
    for j in range (0,size_x):
        temp.append('.')
    for k in range (randint(1,2)):  
        mine=randint(-2,size_x+1)
        if mine>=0 and mine<size_x:
            temp[mine]='*'
    mat.append(temp)

print ('<Input>')
create_mine(mat)

check_mat(mat)
print ('<Output>')
create_mine(mat)
```

```python

from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import requests
import numpy as np
```

```python
if __name__ == "__main__":
    RANK = 100
 
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
    req = requests.get('https://www.melon.com/chart/month/index.htm?classCd=DM0000#params%5Bidx%5D=1&params%5BrankMonth%5D=202005&params%5BisFirstDate%5D=false&params%5BisLastDate%5D=true', headers = header) ## 주간 차트를 크롤링 할 것임
    html = req.text
    parse = BeautifulSoup(html, 'html.parser')
    
 
    titles = parse.find_all("div", {"class": "ellipsis rank01"}) 
    singers = parse.find_all("div", {"class": "ellipsis rank02"}) 
    albums = parse.find_all("div",{"class": "ellipsis rank03"})
    
 
    title = []
    singer = []
    album = []
  
 
    for t in titles:
        title.append(t.find('a').text)
 
    for s in singers:
        singer.append(s.find('span', {"class": "checkEllipsis"}).text)

    for a in albums:
        album.append(a.find('a').text)

 
    for i in range(RANK):
        print('%3d위: %s [ %s ] - %s'%(i+1, title[i], album[i], singer[i]))
```

```python
df = pd.DataFrame({'순위':df.index, '제목': title, '가수':singer, '앨범':album})
df.index = df.index+1
df2 = df.set_index("순위")
df3 = df2[:10]
df3
```
