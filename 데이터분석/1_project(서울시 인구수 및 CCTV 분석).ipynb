# 서울시 CCTV 현황 데이터 분석

##### 1. 서울통계서비스 사이트에서 '주민등록인구(구별)' 통계에서 2018년도 데이터를 다운 받아서 '서울시 인구현황.txt'파일로 저장

##### 2. 파일에서 데이터를 읽어 컬럼이 '자치구', '인구수', '내국인', '외국인', '고령자' 인 데이터 프레임을 만든다.(pop)

```python
import numpy as np
import pandas as pd
```

```python
# 여기서 문제가 delimeter가 ','이 아니라 '\t'이다. 
#그리고 천단위 구분기호가 숫자로 해야함, header행이 여러줄인데 한개만 남게 해야한다. 
#또한 column이 많아 column을 조정 해야한다

pop = pd.read_csv('서울시 인구현황.txt', sep='\t', thousands=',', usecols=[1,3,6,9,13],
                  skiprows=[0,1,2,3], names=['자치구', '인구수', '내국인', '외국인', '고령자'])
pop
```

##### 3. 외국인 비율, 고령자 비율 만들기

```python
pop['외국인 비율'] = pop['외국인']/pop['인구수'] *100
pop['고령자 비율'] = pop['고령자']/pop['인구수'] *100
pop.head() #다볼 필요없이 잘들어 갔나 위에만 확인
```

##### 4.서울시 자치구 년도별 CCTV 설치 현황 데이터를 다운 받아서 '서울시 자치구 년도별 CCTV 설치 현황.xlsx' 파일로 저장한후, 데이터를 읽서 데이터 프레임을 만든다.(cctv)

```python
cctv = pd.read_excel('서울시 자치구 년도별 CCTV 설치 현황.xlsx')
cctv
```

##### 5. '기관명' 컬럼 데이터의 빈칸을 없애고, 컬럼명을 '자치구'로 변경한다.

```python
cctv['기관명'].unique() #.unique는 리스트로 불러오는 방법
```

```python
cctv['기관명'] = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구',
       '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구',
       '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구',
       '은평구', '종로구', '중구', '중랑구']
```

```python
cctv.rename(columns={'기관명':'자치구'}, inplace=True)
cctv.head()
```

##### 6. 2015년 이전의 데이터를 합하여 '2015년 이전' 컬럼을 만들고, 최근 3개년간의 데이터를 합하여 '최근 3개년' 컬럼을 만든 후, '최근 증가율' 컬럼을 만든다.

```python
#NaN 때문에 이렇게 바꿔준다
cctv[['2011년 이전', '2012년', '2013년']] = \
cctv[['2011년 이전', '2012년', '2013년']].fillna(0).astype(int) 

#cctv['2011년 이전'] = ctv['2011년 이전'].fillna(0) .astype(int)
#cctv['2012년'] = ctv['2012년'].fillna(0) .astype(int)
#cctv['2013년'] = ctv['2013년'].fillna(0) .astype(int)
```

```python
cctv.head()
```

```python
cctv['2015년 이전'] =cctv[['2011년 이전', '2012년', '2013년', '2014년', '2015년']].sum(axis=1)
```

```python
cctv.head()
```

```python
#위와 같은 결과 다른방법
cctv['xxx'] = cctv.iloc[:, 2:7].sum(axis=1)
cctv.head()
```

```python
cctv['최근 3개년'] = cctv[['2016년', '2017년', '2018년']].sum(axis=1)
cctv.head()
```

```python
cctv['최근 증가율'] = cctv['최근 3개년']/cctv['2015년 이전']*100
cctv.head()
```

##### 7. '자치구', '소계', '최근 증가율' 컬럼을 제외하고 나머지는 모두 삭제한다.

```python
cctv.drop(['2011년 이전', '2012년', '2013년', '2014년', '2015년'], axis=1, inplace=True)
cctv.head()
```

```python
# 위 방법 말고 del을 사용해도 가능
del cctv['2016년']
del cctv['2017년']
del cctv['2018년']
del cctv['2015년 이전']
del cctv['xxx']
del cctv['최근 3개년']
cctv.head()
```

##### 8. 서울시 인구 데이터와 cctv 데이터를 합친다.

```python
pop.head()
```

```python
df = pd.merge(cctv,pop, on='자치구')
df.head()
```

```python
df.set_index('자치구', inplace=True)
df.head()
```

##### 9. CCTV 현황을 다양한 형태로 시각화하여 보여준다.

```python
# 1.자치구별 CCTV 갯수 hbar
# 2.자치구별 인당 cctv 갯수 hbar
```

```python
import matplotlib as mlp
import matplotlib.pyplot as plt
mlp.rcParams['axes.unicode_minus']=False
mlp.rc('font', family = 'Malgun Gothic')
%matplotlib inline

plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['font.size'] =10
```

```python
# 1.자치구별 CCTV 갯수 (내림차순)

df['소계'].sort_values().plot(kind='barh', grid=True, figsize=(5,5))

plt.title('자치구별 CCTV 갯수')

plt.show()
```

```python
# 2.자치구별 인당 cctv 갯수 내림차순

df['인당 CCTV'] = df['소계']/df['인구수']
df['인당 CCTV'].sort_values().plot(kind='barh', grid=True, figsize=(5,5))
plt.show()
```

```python
# 3. 자치구별 최근 증가율(내림차순)

df['최근 증가율'].sort_values().plot(kind='barh', grid=True, figsize=(5,5))

plt.show()
```
